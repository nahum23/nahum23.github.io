```
from google.colab import files
import io
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objs as go
#import plotly.plotly as py
import seaborn as sns
```

```
#able to upload csv into colab
uploaded1 = files.upload()
```

```
#convert to pandas DF
data = pd.read_csv(io.StringIO(uploaded1['Seasons_Stats.csv'].decode('utf-8')))
```

```
#some names have asteriks so replaced it
data['Player'] = data['Player'].str.replace('*', '')
```

```
#convert all NAs to zeros
data[["Year","Age","G","GS","AST","STL","BLK","PTS","PF","TOV","FT","FTA","2P","2PA","3P","3PA","ORB","TRB","DRB"]] = data[["Year","Age","G","GS","AST","STL","BLK","PTS","PF","TOV","FT","FTA","2P","2PA","3P","3PA","ORB","TRB","DRB"]].fillna(0).astype(int)
```

```
data = data.drop(['B', 'blank2', 'blanl'], axis=1)
```

```
#created per game stats and round to one decimal place
data["PPG"] = data["PTS"] / data["G"]
data["APG"] = data["AST"] / data["G"]
data["RPG"] = data["TRB"] / data["G"]
data["FTPG"] = data["FTA"] / data["G"]
data["TOPG"] = data["TOV"] / data["G"]
data["PPG"] = round(data["PPG"],1)
data["APG"] = round(data["APG"],1)
data["RPG"] = round(data["RPG"],1)
data["FTPG"] = round(data["FTPG"],1)
data["TOPG"] = round(data["TOPG"],1)
```

```
#made three DFs for the years used in this code
eleven = data[(data['Year'] == 2011) & (data['PPG'] >= 17)]
sixteen = data[(data['Year'] == 2016) & (data['PPG'] >= 17)]
seventeen = data[(data['Year'] == 2017) & (data['PPG'] >= 17)]
```
